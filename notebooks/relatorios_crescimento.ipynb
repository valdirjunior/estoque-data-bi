```python
# relatorios_crescimento.ipynb

import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
funcionarios_df = pd.read_csv('../data/extracted/funcionarios.csv')
clientes_fornecedores_df = pd.read_csv('../data/extracted/clientes_fornecedores.csv')
produtos_df = pd.read_csv('../data/extracted/produtos.csv')
movimentacoes_df = pd.read_csv('../data/extracted/movimentacoes.csv')

# Anlise de crescimento no nmero de funcionrios
funcionarios_df['data_admissao'] = pd.to_datetime(funcionarios_df['data_admissao'])
funcionarios_df['ano_admissao'] = funcionarios_df['data_admissao'].dt.year
crescimento_funcionarios = funcionarios_df['ano_admissao'].value_counts().sort_index().reset_index()
crescimento_funcionarios.columns = ['ano', 'num_admissoes']

# Plotar crescimento de funcionrios
plt.figure(figsize=(10, 6))
plt.plot(crescimento_funcionarios['ano'], crescimento_funcionarios['num_admissoes'], marker='o')
plt.title('Crescimento no Nmero de Funcionrios')
plt.xlabel('Ano')
plt.ylabel('Nmero de Admisses')
plt.grid(True)
plt.show()

# Anlise de crescimento na base de clientes e fornecedores
clientes_fornecedores_df['criado_em'] = pd.to_datetime(clientes_fornecedores_df['criado_em'])
clientes_fornecedores_df['ano_criacao'] = clientes_fornecedores_df['criado_em'].dt.year
crescimento_cf = clientes_fornecedores_df['ano_criacao'].value_counts().sort_index().reset_index()
crescimento_cf.columns = ['ano', 'num_cf']

# Plotar crescimento de clientes e fornecedores
plt.figure(figsize=(10, 6))
plt.plot(crescimento_cf['ano'], crescimento_cf['num_cf'], marker='o')
plt.title('Crescimento na Base de Clientes e Fornecedores')
plt.xlabel('Ano')
plt.ylabel('Nmero de Novos Clientes/Fornecedores')
plt.grid(True)
plt.show()

# Anlise de crescimento no nmero de produtos ofertados
produtos_df['criado_em'] = pd.to_datetime(produtos_df['criado_em'])
produtos_df['ano_criacao'] = produtos_df['criado_em'].dt.year
crescimento_produtos = produtos_df['ano_criacao'].value_counts().sort_index().reset_index()
crescimento_produtos.columns = ['ano', 'num_produtos']

# Plotar crescimento de produtos
plt.figure(figsize=(10, 6))
plt.plot(crescimento_produtos['ano'], crescimento_produtos['num_produtos'], marker='o')
plt.title('Crescimento no Nmero de Produtos Ofertados')
plt.xlabel('Ano')
plt.ylabel('Nmero de Novos Produtos')
plt.grid(True)
plt.show()

# Salvar resultados em CSV
crescimento_funcionarios.to_csv('../data/processed/crescimento_funcionarios.csv', index=False)
crescimento_cf.to_csv('../data/processed/crescimento_clientes_fornecedores.csv', index=False)
crescimento_produtos.to_csv('../data/processed/crescimento_produtos.csv', index=False)
```
